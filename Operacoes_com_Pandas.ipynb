```python
df[df['y']>0]
# ATENÇÃO, pedi para filtrar apenas os valores de Y maior que 0 no fata frame
# ou seja na coluna Y tem apenas 2 valores maiores que 0, e mostrou, excluiu linha a,d,e que sao menores que 0 e
#coinsequentemente mostrou os outros valores das outras colunas do dt, mas as da coluna y menores que 0, exclui
```

```python
df[(df['y']>0) & (df['w']>0)]
# analise bem os pares de coclhetes e parenteses
#mostrou apenas a linha que tem em ambas os valores maiores que 0
```

```python
(df['y']>0) & (df['w']>0)
#se tirar o df[] do inicio ele mostra como se fosse booleano nas colunas, ou seja apenas a B possui valores 
#maiores que 0 em y e w
```

```python
df[~(df['y'] > 0) & ~(df['w'] > 0)]
# aqui mostrou com o ~(not) apenas os valores que NÃO SÃO maiores que 0 nas colunas y e w. o mesmo poderia ser usado com o <
```

```python
df[(df['y']<0) & (df['w']<0)]
# mesmo resultado do cod acima, sem usar o ~
```

```python
df.loc[['A', 'B'], ['w']]
#quero valores das linhas a e b da coluna w
```
