```python
esquerda = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
direita = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
esquerda
```

```python
direita
```

```python
pd.concat([esquerda,direita])
```

```python
pd.merge(esquerda, direita, on=['key1', 'key2'])
#organizei de  forma que deixei os keys na frente e os valores das demais colunas nas sequencias
#mas olha, ele usou como base o dr esquerda entao, ele vai unir o que é igual, esqeuda linha 0 é k0,k0, entao ele vai achar na direita uma linha com k0 k0
# pra entao mesclar na linha, logo vem k0 k1, mas no dt da dirita, nao tem k0 k1, então, ele nao coloca esses valores
#depois vem k1 k0, e na direita ele encontra duas linhas com esses valores, entao ele faz uma linha com os valores da esquerda + ou outros que achou na direita
# e repete o valir da esquerda + os segundos valores achados na direita
# logo na esqeurda vem K2 K1, mas nao tem na direita, entao ele exclui essa linha
#DESSA FORMA DE USO MERGE, ELE APENAS MOSTRA AS CHAVES QUE SE ASSEMELHAM NOS DOIS DT, POR EXEMPLO, DUAS BASES DE DADOS QUE EU QUEIRA SABER DETERMINADO PRODRUTO
#EM DETERMINADO MES, CONSIGO TER UMA VISAO USANDO ESSA OPERAÇÃO
```

```python
esquerda
```

```python
direita
```

```python
pd.merge(esquerda, direita, how='outer', on=['key1', 'key2'])
#how seria como eu quero, e com o uso do outer fcaria assim QUERO TODOS OS DADOS DA ESQUERDA E TODOS DA DIREITA, USE O NaN onde for preciso preencher
```

```python
pd.merge(esquerda, direita, how='right', on=['key1', 'key2'])
#exclui a linha 1 e 3 do dataframe esquerda e manteve o dataframe direita intacto
```

```python
direita
```

```python
pd.merge(esquerda, direita, how='left', on=['key1', 'key2'])
#manteve todas as linhas do dataframe esquerda, aumentou a linha 4, pois os valores k1 e k0 repetem no df da direita,mas usou todos os valores da esquerda
# e nao inseirou os valores da linha k2 k0 da direita porque nao tem esse valor na esquerda
```
